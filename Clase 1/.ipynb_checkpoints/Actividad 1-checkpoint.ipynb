```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

```python
df = pd.read_csv("datasets/YPFD.2000.2021.csv", index_col=0)
df["fechaHora"] = pd.to_datetime(df["fechaHora"])
df = df.sort_values(by="fechaHora")
df
```

```python
matplotlib.rcParams['figure.figsize'] = [18,8]
plt.figure()
ax = df.plot(x='fechaHora', y='ultimoPrecio')
ax.set_title("Precio de acciones de YPF", fontsize=16)
ax.set_ylabel("Precio", fontsize=14)
ax.set_xlabel("Fecha", fontsize=14)
plt.legend(loc='best')
plt.show()
```

Los precios de las acciones de YPF se encuentran en cero hasta 2018 donde se observa un pico que indicaría que YPF comienza a operar en bolsa. A simple vista la serie temporal no parece ser estacionaria ni estacional. A comienzos de 2020 se observa una fuerte caído del precio de las acciones probablemente por efecto de la pandemia de Covid-19: el aislamiento de las personas en sus casas y por ende la menor cantidad de vehículos en circulación, provocaron un menor consumo de gasolina. Hacia 2021 se alcanza su máximo histórico dentro del rango de tiempo registrado.

```python
df = df[df.ultimoPrecio > 0]
matplotlib.rcParams['figure.figsize'] = [16,5]
plt.figure()
ax = df.plot(x='fechaHora', y='ultimoPrecio')
ax.set_title("Precio de acciones de YPF", fontsize=16)
ax.set_ylabel("Precio", fontsize=14)
ax.set_xlabel("Fecha", fontsize=14)
plt.legend(loc='best')
plt.show()
```

Se comprueba que las acciones de YPF toman un valor mayor a 0 a partir de 2017

```python
df = pd.read_csv("datasets/TECO2.2010.2021.csv", index_col=0)
df["fechaHora"] = pd.to_datetime(df["fechaHora"])
df = df.sort_values(by="fechaHora")
df
```

```python
df = pd.read_csv("datasets/TECO2.2010.2021.csv", index_col=0).sort_values(by="fechaHora")
df.fechaHora = pd.to_datetime(df["fechaHora"])
matplotlib.rcParams['figure.figsize'] = [18,8]
plt.figure()
ax = df.plot(x='fechaHora', y='ultimoPrecio')
ax.xaxis.set_major_locator(mdates.MonthLocator(1))
ax.set_title("Precio de acciones de Telecom", fontsize=16)
ax.set_ylabel("Precio", fontsize=14)
ax.set_xlabel("Fecha", fontsize=14)
plt.legend(loc='best')
plt.show()
```

Los precios de las acciones de Telecom: a simple vista la serie temporal no parece ser estacionaria ni estacional. Tiene una tendencia creciente. A partir de 2018 se observa una mayor varianza en la serie temporal, alcanzándose un pico histórico (en el rango de tiempo registrado) hacia fines de 2020, con una pronunciada caída a mediados de 2021.

```python
df = pd.read_csv("datasets/BYMA.csv", index_col=0)
df["fechaHora"] = pd.to_datetime(df["fechaHora"])
df = df.sort_values(by="fechaHora")
df
```

```python
df = pd.read_csv("datasets/BYMA.csv", index_col=0).sort_values(by="fechaHora")
df.fechaHora = pd.to_datetime(df["fechaHora"])
matplotlib.rcParams['figure.figsize'] = [18,8]
plt.figure()
ax = df.plot(x='fechaHora', y='ultimoPrecio')
ax.xaxis.set_major_locator(mdates.MonthLocator(1))
ax.set_title("Precio de acciones BYMA", fontsize=16)
ax.set_ylabel("Precio", fontsize=14)
ax.set_xlabel("Fecha", fontsize=14)
plt.legend(loc='best')
plt.show()
```

Por lo que intuyo lo que se muestra es un registro del índice Merval de la bolsa de valores de Argentina. La serie por lo visto no es ni estacional ni estacionaria, presenta un claro descenso hacia 2020 probablemente por la pandemia de Covid-19 y la incertidumbre financiera vivida en ese momento, y presenta un máximo histórico a principios de 2021.
